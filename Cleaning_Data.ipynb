```python
import pandas as pd

df = pd.read_csv("data.tsv", sep="\t")
df.head()
```

```python
# Keep only the relevant columns
df = df[['clue_value', 'category', 'question', 'answer']]

# Drop rows with missing values in those columns
df.dropna(inplace=True)
df
```

```python
# Remove rows where clue_value is 0
df = df[df['clue_value'] != 0]
df
```

```python
df['clue_value'].unique()
```

```python
# Strip whitespace
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

len(df)
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)
len(df)
```

```python
# Save cleaned data
df.to_csv("jeopardy_cleaned.csv", index=False)
```
